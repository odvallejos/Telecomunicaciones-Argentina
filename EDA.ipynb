# Conectividad al servicio de Internet

Consulta las tecnologías disponibles en tu localidad para acceder al servicio de Internet fijo y móvil

```python
df_conect = pd.read_csv("./datasets/Conectividad-servicio-Internet.csv")
```

```python
df_conect.info()
```

```python
df_conect.shape
```

```python
df_conect.head()
```

```python
df_conect.tail()
```

```python
# Elimino las columnas Partido, link, Latitud, longitud porque considero que no son relevantes para el análisis
df_conect.drop(columns=['Partido', 'link', 'Latitud', 'Longitud'], inplace=True)
```

```python
df_conect.head()
```

```python
df_conect.Provincia.unique().shape[0]
```

```python
# Transformo los datos string en numéricos para los tipos de conexión

df_conect['ADSL'] = df_conect['ADSL'].str.replace('SI','1')
df_conect['ADSL'] = df_conect['ADSL'].str.replace('--','0').astype(int)
```

```python
# Transformo los datos string en numéricos para los tipos de conexión
columns = ['ADSL', 'CABLEMODEM', 'DIALUP', 'FIBRAOPTICA', 'SATELITAL', 'WIRELESS', 'TELEFONIAFIJA', '3G', '4G']
for col in columns:
    df_conect[col] = df_conect[col].str.replace('SI','1')
    df_conect[col] = df_conect[col].str.replace('--','0').astype(int)
```

```python
df_conect.info()
```

```python
df_conect.head()
```

```python
total_localidades = df_conect.shape[0]
```

```python
df_conect[( df_conect['ADSL'] + df_conect['CABLEMODEM'] + df_conect['DIALUP'] + df_conect['FIBRAOPTICA'] + df_conect['SATELITAL'] + df_conect['WIRELESS'] + df_conect['TELEFONIAFIJA'] + df_conect['3G'] + df_conect['4G']) == 0 ]
```

```python
df_conect_sin_con = df_conect[( df_conect['ADSL'] + df_conect['CABLEMODEM'] + df_conect['DIALUP'] + df_conect['FIBRAOPTICA'] + df_conect['SATELITAL'] + df_conect['WIRELESS'] + df_conect['TELEFONIAFIJA'] + df_conect['3G'] + df_conect['4G']) == 0 ]
```

```python
total_sin_con = df_conect_sin_con.shape[0]
```

```python
print(f"El {round(total_sin_con/total_localidades*100,2)}% de las localidades no tiene conexión a internet, {total_sin_con} localidades")
```

```python
df_conect_sin_con.Provincia.value_counts()
```

```python
df_conect_sin_con.Provincia.unique().shape[0]
```
